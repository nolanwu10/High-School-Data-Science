```python
import seaborn as sns
import matplotlib.pyplot as plt


sns.set_style("whitegrid")

f, axes = plt.subplots(2, 3, figsize=(15, 10), sharex=False)
sns.despine(left=True)


sns.histplot(poke_df['hp'], ax=axes[0, 0], kde=True, color="b", bins=30).set_title('Distribution of HP')
sns.histplot(poke_df['atk'], ax=axes[0, 1], kde=True, color="r", bins=30).set_title('Distribution of ATK')
sns.histplot(poke_df['def'], ax=axes[0, 2], kde=True, color="g", bins=30).set_title('Distribution of DEF')
sns.histplot(poke_df['spatk'], ax=axes[1, 0], kde=True, color="y", bins=30).set_title('Distribution of SPATK')
sns.histplot(poke_df['spdef'], ax=axes[1, 1], kde=True, color="c", bins=30).set_title('Distribution of SPDEF')
sns.histplot(poke_df['speed'], ax=axes[1, 2], kde=True, color="m", bins=30).set_title('Distribution of SPEED')


plt.tight_layout()
plt.show()
```

```python
type_df = pd.read_csv('/content/drive/MyDrive/Machine Learning/Data/typing_chart.csv')
```

```python
type_long_df = type_df.melt(id_vars=["Types"],
                            var_name="Defending_Type",
                            value_name="Effectiveness")

# Rename the "Types" column for clarity
type_long_df = type_long_df.rename(columns={"Types": "Attacking_Type"})

type_long_df.head()
```

```python

super_effective_df = type_long_df[type_long_df["Effectiveness"] == 2.0]


type_advantage_counts = super_effective_df['Attacking_Type'].value_counts()


plt.figure(figsize=(14, 8))
sns.barplot(x=type_advantage_counts.index, y=type_advantage_counts.values, palette="viridis")
plt.title('Distribution of Type Advantages for Each Primary Type')
plt.xlabel('Primary Type')
plt.ylabel('Number of Types It Is Super Effective Against')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python

basic_pokemon = poke_df[poke_df['evolves_from'] == 'nothing']['name'].unique()


first_stage_pokemon = poke_df[poke_df['evolves_from'].isin(basic_pokemon)]['name'].unique()


second_stage_pokemon = poke_df[poke_df['evolves_from'].isin(first_stage_pokemon)]['name'].unique()


stage_counts = {
    'Basic PokÃ©mon': len(basic_pokemon),
    'First Stage PokÃ©mon': len(first_stage_pokemon),
    'Second Stage PokÃ©mon': len(second_stage_pokemon)
}


plt.figure(figsize=(10, 6))
sns.barplot(x=list(stage_counts.keys()), y=list(stage_counts.values()), palette="pastel")
plt.title('Distribution of PokÃ©mon by Evolutionary Stage')
plt.ylabel('Number of PokÃ©mon')
plt.xlabel('Evolutionary Stage')
plt.tight_layout()
plt.show()
```

## BMI

```python


poke_df['height_m'] = poke_df['height'] * 0.1


poke_df['weight_kg'] = poke_df['weight'] / 10


poke_df['BMI'] = poke_df['weight_kg'] / (poke_df['height_m'] ** 2)

```

```python
highest_bmi_pokemon = poke_df[poke_df['BMI'] == poke_df['BMI'].max()]['name'].iloc[0]
print(highest_bmi_pokemon)
```

```python
bmi_min = poke_df['BMI'].quantile(0.01)  # 1st percentile
bmi_max = poke_df['BMI'].quantile(0.99)  # 99th percentile

# Plot the correlation between BMI and HP with adjusted x-axis limits
plt.figure(figsize=(12, 8))
sns.scatterplot(x=poke_df['BMI'], y=poke_df['hp'], alpha=0.5, color='blue')
plt.title('Correlation between BMI and HP of PokÃ©mon')
plt.xlabel('BMI')
plt.ylabel('HP')
plt.xlim(bmi_min, bmi_max)  # Adjust x-axis limits
plt.tight_layout()
plt.show()
```

```python

poke_df[['ability_1', 'ability_2']] = poke_df['abilities'].str.split(' ', expand=True, n=1)

```

```python
poke_df
```

```python

ability1_stats = poke_df.groupby('ability_1')['total'].mean().sort_values(ascending=False)


ability2_stats = poke_df.groupby('ability_2')['total'].mean().sort_values(ascending=False)


combined_ability_stats = (ability1_stats.add(ability2_stats, fill_value=0) / 2).sort_values(ascending=False)

combined_ability_stats
```

```python
dual_type_df = poke_df[poke_df['type2'] != 'None']


dual_type_df['type_combination'] = dual_type_df['type1'] + '-' + dual_type_df['type2']


dual_type_combination_counts = dual_type_df['type_combination'].value_counts()


top_dual_type_combinations = dual_type_combination_counts.head(10)


plt.figure(figsize=(14, 8))
top_dual_type_combinations.plot(kind='bar', color='lightblue')
plt.title('Top 10 Most Common Dual-Type PokÃ©mon Combinations')
plt.ylabel('Number of PokÃ©mon')
plt.xlabel('Type Combination')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python

type_combination_effectiveness = dual_type_df.groupby('type_combination')['total'].mean().sort_values(ascending=False)


top_effective_type_combinations = type_combination_effectiveness.head(10)


plt.figure(figsize=(14, 8))
top_effective_type_combinations.plot(kind='bar', color='lightgreen')
plt.title('Top 10 Most Effective Dual-Type PokÃ©mon Combinations (Based on Average Total Stats)')
plt.ylabel('Average Total Stats')
plt.xlabel('Type Combination')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
