```python
date_pattern = r'\d{1,2}/\d{1,2}/\d{4} \d{2}:\d{2}'


df = df[df['datetime'].str.match(date_pattern)]

df['datetime'] = pd.to_datetime(df['datetime'], format='%m/%d/%Y %H:%M')
```

```python
df['year'] = df['datetime'].dt.year
df['month'] = df['datetime'].dt.month
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(12, 6))
plt.plot(yearly_counts.index, yearly_counts.values, marker='o', linestyle='-')
plt.xlabel('Year')
plt.ylabel('Number of Sightings')
plt.title('UFO Sightings Over Time')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Filter the dataset for sightings in the United States (US)
us_sightings = df[df['country'] == 'us']

# Filter the dataset for sightings in Canada (CA)
ca_sightings = df[df['country'] == 'ca']

# Filter the dataset for sightings in the United Kingdom (GB)
gb_sightings = df[df['country'] == 'gb']

# Group the data by year and count sightings
us_sightings_by_year = us_sightings['datetime'].dt.year.value_counts().sort_index()
ca_sightings_by_year = ca_sightings['datetime'].dt.year.value_counts().sort_index()
gb_sightings_by_year = gb_sightings['datetime'].dt.year.value_counts().sort_index()

# Create line plots for each country
plt.figure(figsize=(12, 6))
plt.plot(us_sightings_by_year.index, us_sightings_by_year.values, label='US', marker='o', linestyle='-')
plt.plot(ca_sightings_by_year.index, ca_sightings_by_year.values, label='Canada', marker='o', linestyle='-')
plt.plot(gb_sightings_by_year.index, gb_sightings_by_year.values, label='UK', marker='o', linestyle='-')

plt.xlabel('Year')
plt.ylabel('Number of Sightings')
plt.title('UFO Sightings Over Time (US, Canada, UK)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Create subplots with three separate graphs
fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(12, 15), sharex=True)

# Filter the dataset for sightings in the United States (US)
us_sightings = df[df['country'] == 'us']

# Filter the dataset for sightings in Canada (CA)
ca_sightings = df[df['country'] == 'ca']

# Filter the dataset for sightings in the United Kingdom (GB)
gb_sightings = df[df['country'] == 'gb']

# Group the data by year and count sightings for each country
us_sightings_by_year = us_sightings['datetime'].dt.year.value_counts().sort_index()
ca_sightings_by_year = ca_sightings['datetime'].dt.year.value_counts().sort_index()
gb_sightings_by_year = gb_sightings['datetime'].dt.year.value_counts().sort_index()

# Plot the first subplot (US) with a blue line
ax1.plot(us_sightings_by_year.index, us_sightings_by_year.values, label='US', marker='o', linestyle='-', color='blue')
ax1.set_ylabel('Number of Sightings')
ax1.set_title('UFO Sightings Over Time (US)')
ax1.legend()
ax1.grid(True)

# Plot the second subplot (Canada) with a green line
ax2.plot(ca_sightings_by_year.index, ca_sightings_by_year.values, label='Canada', marker='o', linestyle='-', color='green')
ax2.set_ylabel('Number of Sightings')
ax2.set_title('UFO Sightings Over Time (Canada)')
ax2.legend()
ax2.grid(True)

# Plot the third subplot (United Kingdom) with a red line
ax3.plot(gb_sightings_by_year.index, gb_sightings_by_year.values, label='UK', marker='o', linestyle='-', color='red')
ax3.set_xlabel('Year')
ax3.set_ylabel('Number of Sightings')
ax3.set_title('UFO Sightings Over Time (United Kingdom)')
ax3.legend()
ax3.grid(True)

# Adjust spacing between subplots
plt.tight_layout()

# Show the subplots
plt.show()
```

```python
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature

us_df = df[df['country'] == 'us']

fig = plt.figure(figsize=(12, 8))
ax = plt.axes(projection=ccrs.PlateCarree())

plt.scatter(us_df['longitude'], us_df['latitude'], marker='o', color='red', s=3, alpha=0.5, transform=ccrs.PlateCarree())


ax.coastlines()
ax.add_feature(cfeature.BORDERS, linestyle='--')
ax.add_feature(cfeature.STATES.with_scale('10m'))

plt.title('UFO Sightings in the United States')
plt.show()
```

```python
df['daytime'] = df['datetime'].apply(lambda x: 'Day' if 6 <= x.hour < 18 else 'Night')


avg_duration_day_night = df.groupby('daytime')['duration (seconds)'].mean()

plt.figure(figsize=(8, 5))
avg_duration_day_night.plot(kind='bar')
plt.xlabel('Daytime/Nighttime')
plt.ylabel('Average Duration (seconds)')
plt.title('Average UFO Sighting Duration by Daytime/Nighttime')
plt.xticks(rotation=0)
plt.show()
```

```python

most_recent_sighting = df['datetime'].max()

print("Most Recent Sighting Date and Time:", most_recent_sighting)
```

```python

sightings_2014 = df[df['year'] == 2014]

sightings_up_to_current_date = len(sightings_2014)

days_up_to_current_date = (most_recent_sighting - pd.to_datetime('2014-01-01')).days + 1

sighting_ratio = sightings_up_to_current_date / days_up_to_current_date

total_days_2014 = 365


predicted_sightings_2014 = sighting_ratio * total_days_2014

print("Predicted Total Number of Sightings for 2014:", predicted_sightings_2014)
```
